# **Task 1: Lists**
1. Create a list of 10 integers.
2. Find and print the sum, maximum, and minimum values.
3. Add a new element to the list and remove an existing one.
4. Sort the list in ascending and descending order.


| **Aspect**             | **Details**                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
| ---------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition**         | A **list** is a built-in Python data structure used to store **multiple items** in a **single variable**.                                                                                                                                                                                                                                                                                                                                                                     |
| **Purpose**            | To hold a collection of items, which can be of **different data types** (integers, strings, floats, etc.).                                                                                                                                                                                                                                                                                                                                                                    |
| **Type of Structure**  | Ordered, mutable (changeable) collection.                                                                                                                                                                                                                                                                                                                                                                                                                                     |
| **Mutability**         | **Mutable** – you can add, remove, or modify elements after creating the list.                                                                                                                                                                                                                                                                                                                                                                                                |
| **Duplicates Allowed** | **Yes** – lists can store duplicate elements.                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Indexing**           | Supports **positive and negative indexing**. <br> Example: `list[0]` → first element, `list[-1]` → last element.                                                                                                                                                                                                                                                                                                                                                              |
| **Ordering**           | **Ordered** – items maintain the order in which they were added.                                                                                                                                                                                                                                                                                                                                                                                                              |
| **Common Operations**  | - Add items using `append()` or `insert()` <br> - Remove items using `remove()` or `pop()` <br> - Sort using `sort()` <br> - Find length using `len()`                                                                                                                                                                                                                                                                                                                        |
                                                                                                                                                                                                                                 |


```python

# Create a list of 10 integers
numbers = [10, 23, 5, 89, 45, 67, 34, 22, 11, 90]
print('Original List:', numbers)

# Sum, max, and min
print('Sum:', sum(numbers))
print('Max:', max(numbers))
print('Min:', min(numbers))

# Add and remove element
numbers.append(50)
numbers.remove(23)
print('Updated List:', numbers)

# Sort ascending and descending
print('Ascending:', sorted(numbers))
print('Descending:', sorted(numbers, reverse=True))
```

# **Task 2: Tuples**
1. Create a tuple with 5 strings (e.g., fruits).
2. Access and print the second and last elements.
3. Try to modify an element and note the error.
4. Convert the tuple into a list and modify it.


| **Aspect**             | **Details**                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        |
| ---------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition**         | A **tuple** is a built-in Python data structure used to store **multiple items** in a **single variable**, similar to a list.                                                                                                                                                                                                                                                                                                                                                                      |
| **Purpose**            | To store a collection of items that **should not be changed** after creation.                                                                                                                                                                                                                                                                                                                                                                                                                      |
| **Type of Structure**  | Ordered, **immutable** (unchangeable) collection.                                                                                                                                                                                                                                                                                                                                                                                                                                                  |
| **Mutability**         | **Immutable** – once a tuple is created, **you cannot add, remove, or modify** its elements.                                                                                                                                                                                                                                                                                                                                                                                                       |
| **Duplicates Allowed** | **Yes** – tuples can contain duplicate elements.                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
| **Indexing**           | Supports **positive and negative indexing**. <br> Example: `tuple[0]` → first element, `tuple[-1]` → last element.                                                                                                                                                                                                                                                                                                                                                                                 |
| **Ordering**           | **Ordered** – items remain in the order in which they were added.                                                                                                                                                                                                                                                                                                                                                                                                                                  |
| **Use Cases**          | - Storing data that should **not change**, such as days of the week or fixed configuration settings. <br> - Returning multiple values from a function. <br> - Used as keys in dictionaries (since tuples are immutable).                                                                                                                                                                                                                                                                           |
| **Common Operations**  | - Access elements using indexing <br> - Find length using `len()` <br> - Check for membership using `in` keyword                                                                                                                                                                                                                                                                                                                                                                                   |
| **Advantages**         | - **Faster than lists** (because they are immutable). <br> - Can be used as **dictionary keys**.                                                                                                                                                                                                                                                                                                                                                                                                   |
| **Limitations**        | - Cannot add, remove, or modify elements.                                                                                                                                                                                                                                                                                                                                                                                                                                                          |


```python
# Create a tuple
fruits = ('apple', 'banana', 'cherry', 'mango', 'grape')
print('Second element:', fruits[1])
print('Last element:', fruits[-1])

# Tuples are immutable, so modification will give an error
# Convert tuple to list
fruit_list = list(fruits)
fruit_list[2] = 'orange'
print('Modified List:', fruit_list)
```

# **Task 3: Sets**
1. Create two sets with some common and unique elements.
2. Perform union, intersection, and difference operations.
3. Add a new element to the set and remove an existing one.
4. Check if a specific element exists in the set.


| **Aspect**             | **Details**                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        |
| ---------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Definition**         | A **set** is a built-in Python data structure used to store **multiple items** in a **single variable**, but **only unique elements** are allowed.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Purpose**            | To store **unordered collections** of **unique data** and perform **mathematical set operations** like union, intersection, and difference.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        |
| **Type of Structure**  | **Unordered**, mutable collection of **unique items**.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             |
| **Mutability**         | **Mutable** – you can **add or remove elements**, but **elements themselves must be immutable** (e.g., numbers, strings, tuples).                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  |
| **Duplicates Allowed** | **No** – duplicate elements are automatically removed.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             |
| **Indexing**           | **Not supported** – sets are unordered and do not support indexing or slicing.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     |
| **Ordering**           | **Unordered** – elements may appear in a random order.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             |
| **Common Operations**  | - `add()` → Add an element <br> - `remove()` → Remove an element (error if not found) <br> - `discard()` → Remove element (no error if not found) <br> - `union()` → Combine two sets <br> - `intersection()` → Find common elements <br> - `difference()` → Find elements present in one set but not the other                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          


```python
# Create sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Union, Intersection, Difference
print('Union:', set_a | set_b)
print('Intersection:', set_a & set_b)
print('Difference:', set_a - set_b)

# Add and remove
set_a.add(10)
set_a.remove(2)
print('Updated Set A:', set_a)

# Check existence
print('Is 5 in set_a?', 5 in set_a)
```

# **Task 4: Dictionaries**
1. Create a dictionary with 5 key-value pairs (e.g., student name and marks).
2. Print all keys and values.
3. Update a value for an existing key.
4. Delete a key-value pair.
5. Check if a specific key exists in the dictionary.


| **Aspect**             | **Details**                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                              |
| ---------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition**         | A **dictionary** is a built-in Python data structure that stores data in **key-value pairs**. Each key must be **unique**, but values can be duplicated.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Purpose**            | To efficiently store and retrieve data using **unique keys** instead of numerical indexes.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               |
| **Type of Structure**  | **Unordered** (ordered from Python 3.7 onwards), **mutable** collection.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Mutability**         | **Mutable** – you can add, remove, and modify key-value pairs after creation.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            |
| **Duplicates Allowed** | **Keys → No**, each key must be unique. <br> **Values → Yes**, values can be duplicated.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Indexing**           | Uses **keys** instead of numerical positions for accessing values.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                       |
| **Ordering**           | From Python 3.7+, dictionaries maintain **insertion order** of items.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    |
| **Common Operations**  | - `dict[key]` → Access value by key <br> - `dict.get(key)` → Access value safely <br> - `dict.keys()` → Get all keys <br> - `dict.values()` → Get all values <br> - `dict.items()` → Get all key-value pairs <br> - `dict.update()` → Update dictionary <br> - `del dict[key]` → Delete key-value pair                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               


```python
# Create dictionary
students = {'Ali': 85, 'Sara': 90, 'Ahmed': 78, 'Zara': 92, 'Omar': 88}

# Print keys and values
print('Keys:', students.keys())
print('Values:', students.values())

# Update value
students['Ali'] = 95
print('Updated Dictionary:', students)

# Delete a key-value pair
del students['Ahmed']
print('After Deletion:', students)

# Check if key exists
print('Is Sara in dictionary?', 'Sara' in students)
```

| **Aspect**                         | **Details**                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      |
| ---------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Definition**                     | **NumPy** (Numerical Python) is a **Python library** used for **scientific and numerical computing**. It provides support for **multi-dimensional arrays** and mathematical functions to operate on these arrays efficiently.                                                                                                                                                                                                                                                                                                                                                                                                                    |
| **Purpose**                        | To handle **large datasets** and perform **mathematical, statistical, and matrix operations** much faster than using Python lists.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               |
| **Type of Library**                | External library – needs to be installed using: <br>`pip install numpy`                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          |
| **Data Structure**                 | **ndarray (N-dimensional array)** – similar to a Python list, but optimized for **numerical data** and **performance**.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          |
| **Mutability**                     | **Mutable** – you can modify array elements after creation.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      |
| **Duplicates Allowed**             | **Yes** – arrays can contain duplicate values.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
| **Indexing**                       | Supports **positive and negative indexing**, similar to lists. <br> Example: `arr[0]` → first element, `arr[-1]` → last element.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| **Ordering**                       | **Ordered** – elements maintain their position in the array.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     |
| **Why Use NumPy Instead of Lists** | - Much **faster and more efficient** for numerical operations. <br> - Uses **less memory**. <br> - Has **built-in mathematical functions** like `mean()`, `sum()`, `std()`.                                                                                                                                                                                                                                                                                                                                                                                                                                                                      |
| **Common Operations**              | - Create arrays using `np.array()` or `np.arange()` <br> - Reshape arrays using `reshape()` <br> - Perform element-wise operations (+, -, \*, /) <br> - Aggregate functions like `mean()`, `median()`, `sum()`, `max()`, `min()` <br> - Matrix operations like `dot()` and `transpose()`                                                                                                                                                                                                                                                                                                                                                         |
| **Use Cases**                      | - **Data analysis** <br> - **Machine learning** <br> - **Scientific research** <br> - **Image processing** <br> - **Mathematical simulations**                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
| **Advantages**                     | - **High performance** due to optimized C backend. <br> - Easy integration with other libraries like **Pandas**, **Matplotlib**, **Scikit-learn**.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               |
| **Limitations**                    | - Only works well with **numerical data**. <br> - Needs to be **installed separately**.                                                                                                                                                                                                                                                                                                                                                                                                                                                                                


```python
import numpy as np

# Create 1D array
arr = np.arange(1, 21)
print('1D Array:', arr)

# Reshape to 4x5 matrix
matrix = arr.reshape(4, 5)
print('4x5 Matrix:\n', matrix)

# Mean, Median, Std
print('Mean:', np.mean(arr))
print('Median:', np.median(arr))
print('Standard Deviation:', np.std(arr))

# Element-wise operations
arr2 = np.arange(21, 41)
print('Addition:', arr + arr2)
print('Multiplication:', arr * arr2)
```

# **Task 6: Matplotlib**
1. Use Matplotlib to create the following plots:
   - A line plot showing numbers from 1 to 10 and their squares.
   - A bar chart showing the marks of 5 students.
   - A pie chart showing the distribution of 3 categories.
2. Label the axes and add titles to the plots.


```python
import matplotlib.pyplot as plt

# Line Plot
x = list(range(1, 11))
y = [i**2 for i in x]
plt.plot(x, y)
plt.title('Numbers vs Squares')
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.show()

# Bar Chart
students = ['Ali', 'Sara', 'Ahmed', 'Zara', 'Omar']
marks = [85, 90, 78, 92, 88]
plt.bar(students, marks)
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

# Pie Chart
categories = ['Category A', 'Category B', 'Category C']
values = [30, 50, 20]
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Category Distribution')
plt.show()
```
