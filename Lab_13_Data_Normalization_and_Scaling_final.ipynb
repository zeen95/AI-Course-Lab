# **Normalization**
Normalization refers to adjusting the values of numeric data to fit within a specific range, typically [0, 1] or [-1, 1]. This process ensures that no particular feature dominates others due to its scale.

# **Implement data normalization and scaling techniques to preprocess datasets.**

To implement data normalization and scaling techniques to preprocess datasets, we can use Python libraries like **scikit-learn**. Below, I will demonstrate how to apply these techniques using a sample dataset. This example includes loading the dataset, applying normalization and scaling, and then showing the results.

image.png

# **Code Implementation**

# **1. Importing Libraries**

image.png

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

# **2. Loading the Dataset**

image.png

```python
# Load the California Housing dataset
california_housing = fetch_california_housing()
data = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
```

# **3. Min-Max Normalization**

image.png

```python
# Apply Min-Max Normalization
min_max_scaler = MinMaxScaler()
data_min_max_scaled = min_max_scaler.fit_transform(data)
data_min_max_scaled = pd.DataFrame(data_min_max_scaled, columns=data.columns)
```

# **4. Standardization (Z-score Normalization)**


image.png

```python
# Apply Standardization
standard_scaler = StandardScaler()
data_standard_scaled = standard_scaler.fit_transform(data)
data_standard_scaled = pd.DataFrame(data_standard_scaled, columns=data.columns)
```

# **Robust Scaling**

image.png

```python
# Apply Robust Scaling
robust_scaler = RobustScaler()
data_robust_scaled = robust_scaler.fit_transform(data)
data_robust_scaled = pd.DataFrame(data_robust_scaled, columns=data.columns)
```

# **6. Displaying Results**

```python
# Display results

####print(): Displays the first few rows of each scaled dataset to the console. This helps in verifying that the scaling has been applied correctly.
print("Min-Max Normalized Data:")
print(data_min_max_scaled.head())

print("\nStandardized Data:")
print(data_standard_scaled.head())

print("\nRobust Scaled Data:")
print(data_robust_scaled.head())
```
