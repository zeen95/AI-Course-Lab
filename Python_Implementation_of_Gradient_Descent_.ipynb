# **Importing Libraries**

**numpy** is imported for numerical operations, especially for handling arrays and matrices.

**matplotlib.pyplot** is imported for plotting graphs.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Data Generation**

**np.random.seed(0):** Sets the random seed to ensure reproducibility of the random numbers.

 **X = 2 * np.random.rand(100, 1):**  Generates 100 random numbers uniformly distributed between 0 and 2.

 **y = 4 + 3 * X + np.random.randn(100, 1):** Creates the output y based on a linear relationship with X plus some Gaussian noise (np.random.randn(100, 1)).

```python
# Generate some sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

# **Adding Intercept Term**

**X_b = np.c_[np.ones((100, 1)), X]:** Adds a column of ones to X to account for the intercept term
0.This results in a 100x2 matrix where the first column is all ones and the second column is X.

```python
# Add the intercept term to X
X_b = np.c_[np.ones((100, 1)), X]
```

# **Gradient Descent Parameters**

**learning_rate = 0.1:** The step size for each iteration of gradient descent.

**n_iterations = 1000:** The number of iterations to run gradient descent.

**m = 100:** The number of training examples.

```python
# Gradient Descent parameters
learning_rate = 0.1
n_iterations = 1000
m = 100
```

# **Initializing Parameters**

image.png

```python
# Initialize theta randomly
theta = np.random.randn(2, 1)
```

# **Gradient Descent Loop**

image.png

```python
# Gradient Descent
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
```

# **Printing and Plotting Results**

```python
print("Theta found by gradient descent:")
print(theta)
```

Prints the final parameters

 found by gradient descent.

image.png

```python
# Plotting the results
plt.plot(X, y, "b.")
plt.plot(X, X_b.dot(theta), "r-")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression with Gradient Descent")
plt.show()
```
