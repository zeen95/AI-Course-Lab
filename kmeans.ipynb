# k means clustering example


```python
from sklearn.cluster import KMeans
import numpy as np
```

## Create some dummy data and its scatter plot 

```python
from matplotlib import pyplot as plt
%matplotlib inline  

X = np.array([[1.5, 2], [2, 1.5], [2, 1],[4, 3],
              [3.5, 4], [3.5, 3]])

plt.scatter(X[:,0],X[:,1])
```

Lets create a kmeans object with k=2 and fit it with the data. 

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

```

 lets show the cluster lables and predict some new data

```python
print('cluster labels: \n', kmeans.labels_)
print('predicted: \n', kmeans.predict([[0, 0], [4, 4]]))
```

Lets assume the iris datase does not have lables and use k-means to cluster these data. 

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:, :2] # we are using only two features
```

```python
kmeans = KMeans(n_clusters=4).fit(X)
```

```python
# show cluster lables 
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)

# show cluser centers in red
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="red")
```
